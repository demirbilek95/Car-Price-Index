```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
from sklearn.cluster import KMeans
from scipy.stats import norm, skew
from scipy import stats
```

```python
data = pd.read_excel('../data/data_merged_filtered.xlsx',index_col=0)
data.head()
```

```python
data4cluster = data.copy(deep = True)
```

```python
silinecekler = ['il','seri','model','kasa_tipi','motor_hacmi_cc','cekis','renk','garanti','plaka_uyruk','kimden','takas','durumu']
for i in silinecekler:
    del data4cluster[i]
```

```python
data4cluster = pd.get_dummies(data4cluster)
```

```python
data4cluster.head(5)
```

```python
data4cluster.info()
```

```python
data4cluster_std = stats.zscore(data4cluster[data4cluster.columns])
```

```python
data4cluster_std
```

```python
data4cluster_std[:,0] = data4cluster_std[:,0]*200
```

```python
kmeans = KMeans(n_clusters = 3,random_state=0)
```

```python
kmeans.fit(data4cluster_std)
```

```python
kmeans.cluster_centers_
```

```python
labels = kmeans.labels_
```

```python
labels = pd.Series(labels)
```

```python
labels.name='label'
```

```python
labels
```

```python
labels.unique()
```

```python
last = pd.concat([data,labels],axis=1)
```

```python
last
```

```python
last['label'].unique()
```
